Missing values

```python
demographic_data_missing = pd.read_csv('data/life_expectancy_and_income_missing.csv')
```

```python
demographic_data_missing.head()
```

```python
demographic_data_missing.describe()
```

```python
demographic_data_missing.tail()
```

```python
demographic_data_missing.info()
```

```python
demographic_data_missing.isna().sum()
```

```python
demographic_data_missing.fillna(0)

# replacing NA's with 0
```

```python
demographic_data_missing.fillna({
    'life_expectancy': demographic_data_missing['life_expectancy'].mean(),
    'income_per_person': demographic_data_missing['income_per_person'].mean()
})

# filling NA's with the mean of the column
```

```python
# replacing with a more accurate value, e.g. the value of the country (so grouping by country first)

demographic_data_missing.groupby('country')['life_expectancy'].transform('mean')

# note - all values in the group take on the same value
```

```python
demographic_data_missing.fillna({
    'life_expectancy': demographic_data_missing.groupby('country')['life_expectancy'].transform('median')
})
```

```python
# an option is also to drop NA's

demographic_data_missing.dropna()
```
