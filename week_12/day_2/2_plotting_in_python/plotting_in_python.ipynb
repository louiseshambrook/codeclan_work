### Plotting in python

```python
import pandas as pd
demo_data = pd.read_csv('data/life_expectancy_and_income.csv')
```

```python
demo_data_2019 = demo_data.loc[demo_data.year == 2019].copy()
demo_data_2019
```

```python
import matplotlib.pyplot as plt

# importing the matplotlib
```

```python
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10
```

```python
plt.scatter(x = 'income_per_person', y = 'life_expectancy', data = demo_data_2019)
```

```python
plt.scatter(x = 'income_per_person', y = 'life_expectancy', data = demo_data_2019)
plt.xlabel('Income per person (£)')
```

```python
plt.scatter(x = 'income_per_person', y = 'life_expectancy', data = demo_data_2019)
plt.xlabel('Income per person (£)')
plt.ylabel('Life Expectancy')
```

```python
plt.scatter(x = 'income_per_person', y = 'life_expectancy', data = demo_data_2019)
plt.xlabel('Income per person (£)')
plt.ylabel('Life Expectancy')
plt.title('Scatterplot of Income per person vs. life expectancy')
```

```python
demo_data_japan = demo_data.loc[demo_data['country'] == 'Japan'].copy()
```

```python
demo_data_japan.head()
```

```python
sns.lineplot(x = 'year', y = 'fertility_rate', data = demo_data_japan)
plt.title('Fertility rate over time for Japan');
```

```python
sns.barplot(x = 'country', y = 'life_expectancy', data = demo_data_2019.head(5))
```

```python
# sorting the plots
# a case of manipulating the data, and sorting this, rather than the plot

ordered_data = demo_data_2019.head(5).sort_values('life_expectancy')['country']
sns.barplot(x = 'country', y = 'life_expectancy',
            data = demo_data_2019.head(5),
           order = ordered_data);
```

```python
last_20 = demo_data.loc[demo_data['year'] > 2000].copy()
```

```python
last_20
```

```python
import numpy as np
```

```python
sns.barplot(x = 'year', y = 'income_per_person', data = last_20,
           estimator = np.mean,
            ci = False
           )
plt.title('Average income per person by year from 2000 to 2019');
```

```python
# flipping the axes
# REMEMBER to change your x and y

sns.barplot(y = 'year', x = 'income_per_person', data = last_20,
           estimator = np.mean,
            ci = False,
            orient = 'h').invert_yaxis()
plt.title('Average income per person by year from 2000 to 2019');
```

```python
# fixes the tick marks on the x axis

sns.barplot(x = 'year', y = 'income_per_person', data = last_20,
           estimator = np.mean,
            ci = False
           )
plt.title('Average income per person by year from 2000 to 2019')
plt.xticks(rotation = 45);
```
